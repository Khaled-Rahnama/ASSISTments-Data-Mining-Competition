```python
import numpy as np
from pandas import DataFrame
from sklearn.cross_validation import train_test_split
import pandas as pd
from pandas import MultiIndex
```

```python
x = DataFrame(np.arange(24).reshape(6,4), index=MultiIndex.from_arrays([[1,1,3,3,2,2], [1,2,1,2,1,2]], names=['itest','seq']), columns = list("abcd"))
app = DataFrame([[32,1,3,7],[12,3,20,15]], index=MultiIndex.from_arrays([[4,4], [1,2]], names=['itest','seq']), columns = list("abcd"))
x = pd.concat([x,app])
x
```

```python
trunc = x.sort_values("c",ascending=False).groupby('itest').head(1)
trunc
```

```python
from keras.preprocessing import sequence
x_padded =[]
for i, seqs in trunc.groupby("itest"):
    x_padded.append(seqs)
x_padded = sequence.pad_sequences(x_padded, 3, dtype='float64', padding='post', truncating="post", value=9999)
x_padded
```

```python
ITEST_ix = pd.Index(np.repeat(sorted(trunc.index.get_level_values(0).unique()), 3), name='ITEST_id')
new_seq_ix = pd.Index(list(range(0,3))* x_padded.shape[0], name="seq_ix")
paddedData = DataFrame(x_padded.reshape(x_padded.shape[0]*x_padded.shape[1], -1), index=[ITEST_ix, new_seq_ix], columns=trunc.columns)
```
